```python
import pandas as pd
import numpy as np
```

```python
# Acquire 0 indexed vocabulary
f = open('./ap/vocab.txt', 'r')
vocabs = []
while True:
    line = f.readline().strip()
    vocabs.append(line)
    if line == 'buffs': break
f.close()

# Acquire data
f = open('./ap/ap.dat', 'r')
data = pd.DataFrame(columns=vocabs)
for _ in range(2246):
    line = f.readline()
    row = [0] * 10473
    for str in line.strip().split(' '):
        if ':' in str:
            str_split = str.split(':')
            row[int(str_split[0])] = int(str_split[1])
    data.loc[len(data)] = row
f.close()
```

```python
data
```

```python
count_array = []

for elem in vocabs:
    count = 0
    unique = 0
    for i in data[elem]:
        if i != 0:
            unique += 1
        count += i
    count_array.append(count)
    print(elem, count, unique)
```

```python
# remove outliers
mean = np.mean(count_array)
std = np.std(count_array)

# Arbitrary Choice, over 1500 under 10 prune
pruned_vocabs = []
for i in range(len(vocabs)):
    if count_array[i] < 1500 and count_array[i] > 10:
        pruned_vocabs.append(vocabs[i])
pruned_vocabs
```

```python
pruned_data = data[pruned_vocabs]
```

```python
def poisson(x, mean):
    return np.exp(-mean) * np.power(mean, x) / np.math.factorial(x)

def gaussian(x, mean, std):
    return 1 / (std * np.sqrt(2 * np.pi)) * np.exp(-np.power(x - mean, 2) / (2 * np.power(std, 2)))

def gamma(x, alpha, beta):
    return np.power(beta, alpha) / np.math.gamma(alpha) * np.power(x, alpha - 1) * np.exp(-beta * x)

def beta(x, alpha, beta):
    return np.power(x, alpha - 1) * np.power(1 - x, beta - 1) / np.math.beta(alpha, beta)

def dirichlet(x, alpha):
    return np.prod(np.power(x, alpha - 1))

```
